#### Exercise Set 1: Array Operations

```python
import numpy as np
import matplotlib.pyplot as plt

# Exercise 1: Image Processing
def image_processing_exercise():
    """
    Create a simulated image and apply various filters
    """
    # Create a sample 2D array representing a grayscale image
    image = np.random.rand(100, 100)
    
    # Tasks:
    # 1. Apply a threshold to create a binary image
    # 2. Create a border around the image
    # 3. Calculate the average intensity in 10x10 patches
    
    # Your solution here:
    binary_image = image > 0.5
    bordered_image = np.pad(image, pad_width=2, mode='constant', constant_values=1)
    
    # Calculate patch averages
    patches = image.reshape(10, 10, 10, 10).mean(axis=(2, 3))
    
    return binary_image, bordered_image, patches

# Exercise 2: Time Series Analysis
def time_series_exercise():
    """
    Work with stock price data simulation
    """
    # Generate sample daily stock prices for a year
    days = 365
    stock_prices = 100 * (1 + np.random.randn(days).cumsum() * 0.02)
    
    # Tasks:
    # 1. Calculate daily returns
    # 2. Find the highest and lowest prices
    # 3. Calculate 7-day moving average
    
    # Solution template
    daily_returns = np.diff(stock_prices) / stock_prices[:-1]
    extremes = {'max': np.max(stock_prices), 'min': np.min(stock_prices)}
    moving_avg = np.convolve(stock_prices, np.ones(7)/7, mode='valid')
    
    return daily_returns, extremes, moving_avg
```

```python
time_series_exercise()
```

#### Exercise Set 2: Broadcasting Rules

```python
# Exercise 3: Multi-dimensional Data Normalization
def normalization_exercise():
    """
    Normalize data across different dimensions using broadcasting
    """
    # Create sample data: batch of images (batch_size, height, width, channels)
    batch_size, height, width, channels = 32, 64, 64, 3
    images = np.random.rand(batch_size, height, width, channels)
    
    # Tasks:
    # 1. Normalize each channel independently
    # 2. Normalize each image in the batch
    # 3. Apply per-batch channel standardization
    
    # Channel normalization (broadcasting across batch, height, width)
    channel_means = images.mean(axis=(0, 1, 2), keepdims=True)
    channel_stds = images.std(axis=(0, 1, 2), keepdims=True)
    channel_normalized = (images - channel_means) / channel_stds
    
    return channel_normalized

# Exercise 4: Feature Scaling in Machine Learning
def feature_scaling_exercise():
    """
    Scale features for machine learning using broadcasting
    """
    # Create sample feature matrix (samples × features)
    X = np.random.randn(100, 5)
    
    # Tasks:
    # 1. Standardize features (zero mean, unit variance)
    # 2. Apply min-max scaling
    # 3. Add polynomial features using broadcasting
    
    # Solutions
    standardized = (X - X.mean(axis=0)) / X.std(axis=0)
    min_max = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
    polynomial = np.column_stack([X, X**2, X**3])
    
    return standardized, min_max, polynomial
```

#### Exercise Set 3: Linear Algebra Operations

```python
# Exercise 5: Portfolio Optimization
def portfolio_optimization_exercise():
    """
    Implement a simple portfolio optimization using linear algebra
    """
    # Generate sample returns and covariance matrix
    n_assets = 4
    returns = np.random.randn(n_assets) * 0.1 + 0.05  # Expected returns
    cov_matrix = np.random.randn(n_assets, n_assets)
    cov_matrix = cov_matrix.T @ cov_matrix  # Ensure positive semi-definite
    
    # Tasks:
    # 1. Find the optimal weights (simplified Markowitz)
    # 2. Calculate portfolio return and risk
    # 3. Implement constraints (weights sum to 1)
    
    # Basic solution (not considering constraints)
    weights = np.linalg.inv(cov_matrix) @ returns
    weights = weights / np.sum(weights)  # Normalize to sum to 1
    
    portfolio_return = np.dot(weights, returns)
    portfolio_risk = np.sqrt(weights @ cov_matrix @ weights)
    
    return weights, portfolio_return, portfolio_risk

# Exercise 6: Image Compression using SVD
def image_compression_exercise():
    """
    Implement image compression using Singular Value Decomposition
    """
    # Create sample image
    image = np.random.rand(50, 50)
    
    # Tasks:
    # 1. Perform SVD
    # 2. Reconstruct image with different numbers of components
    # 3. Calculate compression ratio and error
    
    U, s, Vt = np.linalg.svd(image)
    
    def reconstruct(k):
        return U[:, :k] @ np.diag(s[:k]) @ Vt[:k, :]
    
    reconstructions = {
        'low': reconstruct(5),
        'medium': reconstruct(15),
        'high': reconstruct(30)
    }
    
    errors = {k: np.mean((image - v)**2) for k, v in reconstructions.items()}
    
    return reconstructions, errors
```

#### Bonus Challenge Exercise: Combining All Concepts

```python
def neural_network_exercise():
    """
    Implement a simple neural network forward pass combining all concepts
    """
    # Generate sample data
    batch_size = 32
    input_size = 784  # e.g., MNIST
    hidden_size = 128
    output_size = 10
    
    # Create random input and weights
    X = np.random.randn(batch_size, input_size)
    W1 = np.random.randn(input_size, hidden_size) / np.sqrt(input_size)
    W2 = np.random.randn(hidden_size, output_size) / np.sqrt(hidden_size)
    
    def forward_pass(X, W1, W2):
        # Tasks:
        # 1. Implement layer operations using matrix multiplication
        # 2. Apply activation function (ReLU) using broadcasting
        # 3. Implement batch normalization
        # 4. Calculate softmax output
        
        # First layer
        hidden = X @ W1
        hidden_activated = np.maximum(0, hidden)  # ReLU
        
        # Batch normalization
        hidden_mean = hidden_activated.mean(axis=0, keepdims=True)
        hidden_std = hidden_activated.std(axis=0, keepdims=True)
        hidden_normalized = (hidden_activated - hidden_mean) / (hidden_std + 1e-8)
        
        # Output layer
        logits = hidden_normalized @ W2
        
        # Softmax
        exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))
        probabilities = exp_logits / exp_logits.sum(axis=1, keepdims=True)
        
        return probabilities, hidden_normalized
    
    return forward_pass(X, W1, W2)

# Run and visualize results
probabilities, hidden = neural_network_exercise()
print("Output probabilities shape:", probabilities.shape)
print("Hidden layer activations shape:", hidden.shape)

# Visualize hidden layer activations
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.imshow(hidden[:10, :10], cmap='viridis')
plt.title('First 10 samples × 10 hidden units')
plt.colorbar()

plt.subplot(122)
plt.plot(probabilities[0], 'o-')
plt.title('Output probabilities for first sample')
plt.xlabel('Class')
plt.ylabel('Probability')
plt.tight_layout()
plt.show()
```
